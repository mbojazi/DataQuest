## In Particular, Do Students in More Popular Majors Make More Money?  The Following Scatter Plots Shed Light on the Question.

```python
recent_grads.plot(x='Total', y='Low_wage_jobs', kind='scatter', title='Low_wage_jobs vs. Total', xlim=(-20000,425000), ylim=(-5000,55000))
```

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Median vs. Total', xlim=(-20000,425000), ylim=(-10000,120000))
```

From the plot of Low_wage_jobs vs. Total, we see that, in general, as the total number of people with a certain major (or, major popularity) increases, so does the number of people in low-wage service jobs post graduation.  There is a strong, positive correlation of major popularity with students in such majors ending up in low-wage service jobs.  These increasing numbers of students in low-wage services jobs post graduation with increasing major popularity result in the downward trend of median earnings as said popularity increases in the plot of Median vs. Total.  It appears, then, that students in more popular majors make less money.    

## Do Students that Majored in Subjects that were Majority Female Make More Money?  The Following Scatter Plot Sheds Light on the Question.

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Median vs. ShareWomen', xlim=(-0.05,1.05), ylim=(-10000,120000))
```

From the above plot, there is a clear downward trend in the median earnings as the percentage of women with a particular major increases.  It appears, then, students majoring in subjects increasingly dominated by a female population make less money.
